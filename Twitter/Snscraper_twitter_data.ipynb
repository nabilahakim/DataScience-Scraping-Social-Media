```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Creating list to append tweet data
tweets_list = []
keyword = 'prabowo'
# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweets in enumerate(sntwitter.TwitterSearchScraper(keyword+'').get_items()):
    if i>2500:
        break
    tweets_list.append([tweets.date, tweets.url, tweets.rawContent, tweets.user.username, tweets.user.created,
                         tweets.user.verified, tweets.user.followersCount,tweets.retweetCount, tweets.likeCount, tweets.replyCount, tweets.hashtags])
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Url', 'Text', 'Username','User Created', 'Verified','Followers Count','Retweet Count','Like Count','Reply Count','Hashtags'])
```

```python
tweets_df
```

```python
tweets_df.to_csv(keyword+".csv", index = False)
```
